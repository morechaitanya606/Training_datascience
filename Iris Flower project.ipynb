#### Print the names of iris flowers with the highest and lowest petal lengths.

```python
x = df.sort_values(by='petal_length')
m =x.iloc[0]['species']
n =x.iloc[-1]['species']
print("Flower with the lowest petal length:",m)
print("Flower with the highest petal length:", n)
```

#### Find information about the iris flowers with the widest sepals.

```python
x= df['sepal_width'].max()
m= df[df['sepal_width'] == x]

# Print information about the iris flowers with the widest sepals
print("Information about iris flowers with the widest sepals:")
print(m)
```

#### Plot the distribution of petal lengths for each species.

```python
import matplotlib.pyplot as plt
```

```python
grouped_data = df.groupby('species')['petal_length'].apply(list)

# Plot the distribution of petal lengths for each species
plt.figure(figsize=(8, 6))
plt.hist(grouped_data, bins=10, label=grouped_data.index)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Lengths for Each Species')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```
