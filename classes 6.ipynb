#### NORMALIZATION 

```python
x = np.array([11,22,75,45,12,82,15,45])
```

```python
def normalize(a):
    norm = (a-a.min()) / (a.max() - a.min())
    return norm
```

```python
normalize(x)
```

```python
nx=normalize(x)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual data')
plt.plot(x)
plt.subplot(1,2,2)
plt.title('Normailize Data') 
plt.plot(nx);
```

```python
y = np.array([784,520,410,785,102,147,258,369,999,100])
```

```python
ny= normalize(y)
ny
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual data')
plt.plot(y)
plt.subplot(1,2,2)
plt.title('Normailize Data') 
plt.plot(ny);
```
